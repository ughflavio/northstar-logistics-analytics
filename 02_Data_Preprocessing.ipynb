Loading all Datasets

```python
deliveries = pd.read_csv(
    '/content/northstar_dataset/deliveries.csv'
)

drivers = pd.read_csv(
    '/content/northstar_dataset/drivers.csv'
)

vehicles = pd.read_csv(
    '/content/northstar_dataset/vehicles.csv'
)

incidents = pd.read_csv(
    '/content/northstar_dataset/incidents.csv'
)

complaints = pd.read_csv(
    '/content/northstar_dataset/complaints.csv'
)

customers = pd.read_csv(
    '/content/northstar_dataset/customers.csv'
)

orders = pd.read_csv(
    '/content/northstar_dataset/orders.csv'
)

hubs = pd.read_csv(
    '/content/northstar_dataset/hubs.csv'
)
```

Converting Datetime fields

```python
deliveries['dispatch_time'] = pd.to_datetime(

    deliveries['dispatch_time']

)

deliveries['delivery_completed_at'] = pd.to_datetime(

    deliveries['delivery_completed_at']

)
```

Verifying Datetime Conversions

```python
print(deliveries.dtypes)
```

Creating Delivery Duration KPIs

```python
deliveries['delivery_duration_minutes'] = (

    deliveries['delivery_completed_at']

    -

    deliveries['dispatch_time']

).dt.total_seconds() / 60
```

Verifying KPI creation

```python
print(

    deliveries[
        ['dispatch_time',
         'delivery_completed_at',
         'delivery_duration_minutes']
    ].head()

)
```

Handling Missing Customer Ratings

```python
deliveries['customer_rating_post_delivery'] = (

    deliveries['customer_rating_post_delivery']

    .fillna(

        deliveries[
            'customer_rating_post_delivery'
        ].mean()

    )

)
```

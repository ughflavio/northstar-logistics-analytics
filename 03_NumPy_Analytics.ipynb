NumPy Based Numerical Analytics

This notebook applies NumPy numerical computing techniques to analyse cleaned operational datasets. Vectorised statistical computations are performed to support business intelligence and performance evaluation.

Import Libraries

```python
import pandas as pd
import numpy as np
```

Loading Cleaned datasets

```python
df = pd.read_csv("cleaned_deliveries.csv")
```

Checking Columns

```python
print(df.columns)
```

Cleaning Numerical Column

```python
clean_values = pd.to_numeric(
    df['delivery_duration_minutes'],
    errors='coerce'
).dropna()
```

Converting to Numpy Array

```python
delivery_times = np.array(clean_values)

print(delivery_times[:10])
```

Statistical Analysis

```python
print("Mean:",
      np.mean(delivery_times))

print("Median:",
      np.median(delivery_times))

print("Standard Deviation:",
      np.std(delivery_times))

print("Maximum:",
      np.max(delivery_times))

print("Minimum:",
      np.min(delivery_times))
```

Percentile Analysis

```python
print("25th Percentile:",
      np.percentile(delivery_times, 25))

print("50th Percentile:",
      np.percentile(delivery_times, 50))

print("75th Percentile:",
      np.percentile(delivery_times, 75))
```

Vectorised Filtering

```python
late_deliveries = delivery_times[
    delivery_times > 60
]

print("Late Deliveries:",
      len(late_deliveries))
```

Analytical Interpretation

NumPy vectorised computations were used to evaluate operational performance metrics from the cleaned dataset. Statistical analysis identified average delivery durations, variability patterns, percentile distributions, and delayed operational activities exceeding predefined thresholds.
